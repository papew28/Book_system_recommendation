```python
map_isbn={k:i for i,k in enumerate(books["ISBN"].unique())}
ratings_with_books["unique_book_id"]=ratings_with_books["ISBN"].map(map_isbn)
ratings_with_books.head(5)
```

# collaborative filtering(centré utilisateur)

Pour le systéme de recommandation nous avons utilisé le filtrage collaborative qui va permettre de recommander des livres à des utilisateurs ont au moins déja noté un livre

```python
dataset=ratings_with_books[["User-ID","unique_book_id","Book-Rating","Book-Title"]]
```

```python
user_book_matrix=dataset.pivot(index="User-ID", columns="unique_book_id", values="Book-Rating").fillna(0)
user_book_matrix.head()
```

```python
user_book_matrix.index
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_book_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_book_matrix.index, columns=user_book_matrix.index)
user_similarity_df.head()
```

```python
def get_recommendations(user_id, num_recommendations=5):
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:]
    similar_users_ratings = user_book_matrix.loc[similar_users]
    mean_ratings = similar_users_ratings.apply(lambda row: np.dot(row, user_similarity_df[user_id][similar_users]) / user_similarity_df[user_id][similar_users].sum(), axis=0)
    books_already_rated = user_book_matrix.loc[user_id][user_book_matrix.loc[user_id] > 0].index
    recommendations = mean_ratings.drop(books_already_rated).sort_values(ascending=False).head(num_recommendations)
    index=recommendations.index
    return dataset[dataset["unique_book_id"].isin(index)]["Book-Title"].unique()
get_recommendations(user_id=67, num_recommendations=5)
```

# Content Based System

```python
dataset=ratings_with_books[["User-ID","unique_book_id","Book-Rating","Book-Title","Book-Author","Publisher"]]
dataset.head()
```

```python
dataset.info()
```

Maintenant on va combiner les trois derniers qui vont constituer le contenu sur lequel on va se baser pour faire les recommandations

```python
dataset['content'] = dataset['Book-Title'] + ' ' + dataset['Book-Author'] + ' ' + dataset['Publisher']
```

```python
# je réduis la quantité de données pour ne pas avoir un probléme de mémoire
dataset=dataset.sample(10000,random_state=24)
dataset.reset_index(drop=True, inplace=True)
dataset.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Feature extraction en utilisant  TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(dataset['content'])
print(tfidf_matrix)
```

```python
# on utilise le consinus similarity pour voir à quel point dont livre se resemble basé sur le contenu
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
#cette fonction de recommandation va permettre de recommander 10 livres à un utilisateur
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = dataset[dataset['Book-Title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  
    book_indices = [i[0] for i in sim_scores]
    return dataset['Book-Title'].iloc[book_indices]
```

```python
print(get_recommendations('The Testament'))
```

```python
print(get_recommendations('The Love Hunter'))
```
